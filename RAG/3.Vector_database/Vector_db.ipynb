```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

print("JB")
```

# Step 2 : Create Documents

```python
from langchain_core.documents import Document

doc1=Document(
    page_content="Virat Kohli",
    metadata={"team":"RCB"}
)

doc2=Document(
    page_content="Rohit Sharma",
    metadata={"team":"MI"}
)

doc3=Document(
    page_content="MS Dhoni",
    metadata={"team":"CSK"}
)
```

# Step 3 : Put documents into list

```python
docs=[doc1,doc2,doc3]
```

# Step 4 : Create chroma vector store

```python
vector_store = Chroma(
    embedding_function = OpenAIEmbeddings(),
    persist_directory="my_chroma_db",               # persist_directory = means save everything parmenantly
    collection_name="sample_collection"
)
```

# Step 5 : Add documents

```python
vector_store.add_documents(docs)
```

 # Step 6 : View Everything

```python
vector_store.get(include=["embeddings","documents","metadatas"])
```

# Step 7 : Similarity Search

```python
vector_store.similarity_search(query="who among there are from CSK ?",k=1)
# k means top most similar documents here we ask give top 1 similar
```

# Step 8 : Similarity Score

```python
vector_store.similarity_search_with_score("who among there are from CSK ?")
```
